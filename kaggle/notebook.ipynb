```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/digitaljournal_benchmarks.csv')
print('Dataset shape:', df.shape)
print(df.head(10))
```

```python
signal_cols = ['publication_readiness_score','ai_visibility_score','content_quality_score','discoverability_score','benchmark_score','editorial_standards_score']
signal_avgs = df[signal_cols].mean()
colors = ['#3b82f6','#06b6d4','#10b981','#f59e0b','#8b5cf6','#ef4444']
plt.figure(figsize=(12,5))
plt.bar(signal_avgs.index, signal_avgs.values, color=colors)
plt.title('GetonDigitalJournal IQ - Average Content Intelligence Scores', fontsize=14, fontweight='bold')
plt.xlabel('Content Signal')
plt.ylabel('Average Score (0-100)')
plt.xticks(rotation=20)
plt.axhline(y=70, color='green', linestyle='--', alpha=0.5, label='Healthy threshold')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
platform_cols = ['chatgpt_score','google_ai_score','perplexity_score','gemini_score']
platform_avgs = df[platform_cols].mean()
plt.figure(figsize=(8,5))
plt.bar(platform_avgs.index, platform_avgs.values, color=['#10b981','#3b82f6','#f59e0b','#8b5cf6'])
plt.title('GetonDigitalJournal IQ - AI Platform Visibility Scores', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
